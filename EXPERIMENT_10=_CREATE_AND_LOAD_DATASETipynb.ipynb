```python
import pandas as pd

data = {
    "Roll_no": [1, 2, 3, 4, 5],
    "Gender": ["male", "female", "male", "female", "male"],
    "Department": ["ENTC", "CS", "IT", "AI ML", "ELECTRICAL"],
    "CGPA": [8.5, 9.2, 7.8, 9.5, 8.9]
}

df = pd.DataFrame(data)

df.to_excel("student_data.xlsx", index=False)
```

```python
#access the informarion about Department column
df["Department"]

```

```python
# Access information of roll no 3

roll_no_3 = df[df["Roll_no"] == 3]
print(roll_no_3)
```

```python
# Find out data types of each column

df.dtypes
```

```python
# Perform mean,median and mode operation

print("Mean:", df["CGPA"].mean())
print("Median:", df["CGPA"].median())
print("Mode:", df["CGPA"].mode()[0])
```

```python
# Insert a new column grade

df["Grade"] = ["A", "A+", "B", "A+", "A"]

print(df)
```

```python
# Delete Gender column

df = df.drop("Gender", axis=1)

print(df)
```
